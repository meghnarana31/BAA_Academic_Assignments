```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['price'], bins = 8)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Selling Price')

# the highest number of cars sell in the 5000 - 10000 price bracket
```

```python
plt.scatter(df['enginesize'], df['price'])
plt.xlabel('Engine Size (cc)')
plt.ylabel('Price')
plt.title('Engine Size vs Selling Price')
plt.show()

# with an increase in engine size, the price of the car increases 
```

```python
import seaborn as sns
```

```python
sns.boxplot(x='carbody', y='price', data=df)
plt.xlabel('Car Type')
plt.ylabel('Price')
plt.title('Distribution of Selling Price by Car Type')
plt.show()

# The highest median selling price is observed for convertible cars, while the lowest median selling price is observed for hatchback cars. 
# The interquartile ranges (IQRs) also vary across different car body types, with the largest IQR observed for the hardtop cars.
# There are also some outliers in the data, particularly for the sedan and hatchback car types, indicating that there are some cars that have much higher selling prices compared to the others in their respective categories.
```

```python
sns.violinplot(x='carbody', y='price', hue='fueltype', data=df, split=True)
plt.xlabel('Car Body Type')
plt.ylabel('Price')
plt.title('Distribution of Selling Price by Car Body Type and Fuel Type')
plt.show()
```

```python
sns.pairplot(df, vars=['price', 'enginesize', 'horsepower', 'curbweight'], hue='fueltype')
plt.title('Pairplot of Price, Engine Size, Horsepower, and Curb Weight')
plt.show()

# The plot suggests that there are positive correlations between price, engine size, horsepower, and curb weight, regardless of the fuel type.
# For both gas and diesel cars, there is a clear positive relationship between price and each of the other variables, indicating that these variables may be important predictors of selling price.
# The plot also shows that the relationships between these variables tend to be similar for gas and diesel cars, indicating that the relationships between the variables are not strongly affected by the fuel type.
```

```python
sns.barplot(x='carbody', y='price', hue='fueltype', data=df, ci=None)
plt.xlabel('Car Body Type')
plt.ylabel('Average Price')
plt.title('Average Selling Price by Car Body Type and Fuel Type')
plt.show()

# For gas cars, the average selling price is highest for convertibles, followed by hardtop, sedans, and wagons. However, for diesel cars, the average selling price is highest for wagons, followed by sedans.
# The plot also suggests that, on average, diesel cars tend to be more expensive than gas cars across the two categories.
```
