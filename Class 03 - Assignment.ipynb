```python
top_players = df.sort_values(by='overall', ascending=False).head(10)
(top_players[['short_name', 'age', 'nationality_name', 'club_name', 'overall']])
```

```python
colors = ['pink', 'blue']
explode = (0.05, 0.05)

df['preferred_foot'].value_counts().plot(kind='pie', autopct='%1.0f%%', colors=colors, explode=explode, shadow=True)
plt.title("Distribution of Preferred Foot")
```

```python
plt.figure(figsize = (18, 8))
ax = sns.countplot(x ='age', data = df)
ax.set_xlabel(xlabel = 'Age', fontsize = 16)
ax.set_ylabel(ylabel = 'Count of Players', fontsize = 16)
ax.set_title(label = 'Count of Age', fontsize = 20)
for i in ax.containers:
    ax.bar_label(i)
```

```python
Pos = df['player_positions'].value_counts().head(10)
plt.figure(figsize = (20, 9))
ax = sns.barplot(x=Pos.index,y= Pos,)
ax.set_xlabel(xlabel = 'Positions', fontsize = 16)
ax.set_ylabel(ylabel = 'Count', fontsize = 16)
ax.set_title(label = 'Count of Player Positions', fontsize = 20)  
for i in ax.containers:
    ax.bar_label(i)
```

```python
plt.figure(figsize = (20, 9))
ax = sns.countplot(x ='body_type', data = df, palette='bright')
ax.set_xlabel(xlabel = 'Body Type', fontsize = 17)
ax.set_ylabel(ylabel = 'Count', fontsize = 17)
ax.set_title(label = 'Count of Body Type', fontsize = 20) 
for i in ax.containers:
    ax.bar_label(i)
```

```python
plt.figure(figsize = (20, 9))
ax = sns.countplot(y ='work_rate', data = df)
ax.set_xlabel(xlabel = 'Work Rate', fontsize = 17)
ax.set_ylabel(ylabel = 'Count', fontsize = 17)
ax.set_title(label = 'Count of Work Rate', fontsize = 20)
for i in ax.containers:
    ax.bar_label(i)
```

```python
sns.relplot(data=df,x="potential", y="wage_eur", hue="preferred_foot", col="international_reputation")
```

```python
top_countries = df['nationality_name'].value_counts().head(5)
top_countries_name = top_countries.index
df_country_age = df.loc[df['nationality_name'].isin(top_countries_name)& df['age']]
plt.figure(1 , figsize = (12,6))
ax = sns.boxplot(x = 'nationality_name' , y = 'age' , data = df_country_age, palette='spring')
ax.set_xlabel(xlabel = 'Nationality', fontsize = 17)
ax.set_ylabel(ylabel = 'Age', fontsize = 17)
plt.title('Age wise Distribution in top countries')
```
